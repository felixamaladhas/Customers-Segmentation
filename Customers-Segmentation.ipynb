```python
sns.set_style("white")
plt.figure(figsize=(10, 6))
plt.boxplot([dataset[dataset['Gender'] == 'Male']['Spending Score (1-100)'],
             dataset[dataset['Gender'] == 'Female']['Spending Score (1-100)']])
plt.xticks([1, 2], ['Male', 'Female'])
plt.xlabel('Gender')
plt.ylabel('Spend score')
plt.title('Gender vs Spend score')
plt.show()
```

```python
import scipy.stats as stats

male_scores = dataset[dataset['Gender'] == 'Male']['Spending Score (1-100)'] #subset data to get Male Spend_score
female_scores = dataset[dataset['Gender'] == 'Female']['Spending Score (1-100)'] #subset data to get Female Spend_score

# Perform independent t-test
t_statistic, p_value = stats.ttest_ind(male_scores, female_scores)

alpha = 0.05  # significance level

print("t-statistic:", t_statistic)
print("p-value:", p_value)

if p_value < alpha:
    print("Reject null hypothesis. There is a significant difference between male and female scores.")
else:
    print("Fail to reject null hypothesis. There is no significant difference between male and female scores.")
```

```python
male_scores = dataset[dataset['Gender'] == 'Male']['Age'] #subset data to get Male Age
female_scores = dataset[dataset['Gender'] == 'Female']['Age'] #subset data to get Female Age

# Perform independent t-test
t_statistic, p_value = stats.ttest_ind(male_scores, female_scores)

alpha = 0.05  # significance level

print("t-statistic:", t_statistic)
print("p-value:", p_value)

if p_value < alpha:
    print("Reject null hypothesis. There is a significant difference between male and female scores.")
else:
    print("Fail to reject null hypothesis. There is no significant difference between male and female scores.")
```

```python
## Scaling 

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(dataset[['Annual Income (k$)']])
dataset['Annual Income (k$)'] = scaler.transform(dataset[['Annual Income (k$)']])
```

```python
scaler.fit(dataset[['Spending Score (1-100)']])
dataset['Spending Score (1-100)'] = scaler.transform(dataset[['Spending Score (1-100)']])
```

```python
## uisng 3rd and 4th feature only and converting them into numpy array

x = dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
plt.scatter(dataset['Annual Income (k$)'],dataset['Spending Score (1-100)'])
```

```python
## Elbow Method

from sklearn.cluster import KMeans
# Within Cluster Sum of Squares = wcss
wcss=[]
for i in range(1,21):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
k_range = range(1,21)
plt.plot(k_range,wcss)
plt.title('Elbow Method')
plt.xlabel('k_range')
plt.ylabel('wcss')
plt.show()
```

```python
k_mean = KMeans(n_clusters=5)
k_mean
```

```python
Label_or_predicted = k_mean.fit_predict(x)
```

```python
Label_or_predicted
```

```python
dataset['cluster'] = Label_or_predicted
```
